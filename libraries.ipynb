```python
print("Mean value og cgpa",df['cgpa'].mean())
print("Std value og cgpa",df['cgpa'].std())
print("Min value og cgpa",df['cgpa'].min())
print("Max value og cgpa",df['cgpa'].max())
```

```python
# finding the boundary values (empirical formula)
# value below 5.11 and above 8.80 are outliers
print("Highest allowed",df['cgpa'].mean()+3*df['cgpa'].std())
print("Lowest allowed",df['cgpa'].mean()-3*df['cgpa'].std())
```

```python
#TRIMMING (in our data 5 outliers r there)
df[(df['cgpa']>8.80)|(df['cgpa']<5.11)]
```

```python
# remove the outliers
new_df=df[(df['cgpa']<8.80)&(df['cgpa']>5.11)]
new_df
```

```python
# formula of zscore (value between -3 and +3)
# formula x-meu/sigma
df['cgpa_zscore']=(df['cgpa']-df['cgpa'].mean())/df['cgpa'].std()
```

```python
df[df['cgpa_zscore']>3]
```

```python
df[df['cgpa_zscore']<-3]
```

```python
df[(df['cgpa_zscore']>3)|(df['cgpa_zscore']<-3)]
```

```python
# trimming
new_df=df[(df['cgpa_zscore']<3)&(df['cgpa_zscore']>-3)]
print(new_df)
```

```python
upper_limit=df['cgpa'].mean()+3*df['cgpa'].std()
lower_limit=df['cgpa'].mean()-3*df['cgpa'].std()
print(upper_limit)
print(lower_limit)
```

```python

 # replacing outliers with uppe limit
 df['cgpa']=np.where(
    df['cgpa']>upper_limit,
    upper_limit,
    np.where(
        df['cgpa']<lower_limit,
        lower_limit,
        df['cgpa']
    )
)
 df
```
